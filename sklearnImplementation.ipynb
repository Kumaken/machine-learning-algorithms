```python
from sklearn.neural_network import MLPClassifier as MLP
from sklearn.datasets import load_iris
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data = pd.read_csv("tennis.csv")
iris = load_iris()
print(iris)
```

```python
clf = MLP(activation='logistic', solver='adam', learning_rate='constant', learning_rate_init=0.1, max_iter=200, shuffle=True, verbose=True)
```

```python
iris_tree = clf.fit(iris.data, iris.target)
```

```python
clf.predict([[5.9, 3. , 5.1, 1.8]])
```
